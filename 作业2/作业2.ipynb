### 118A. String Task

implementation/strings, 1000, http://codeforces.com/problemset/problem/118/A







```python
s=input()
l=[]
a=["A","E","I","O","U",'Y','a','e','i','o','u','y']
for i in range(len(s)):
    if s[i] not in a:
        b=s[i]
        if ord("A")<=ord(s[i])<=ord('Z'):
            b=chr(ord(s[i])-ord("A")+ord('a'))
        l.append(b)
for i in range(len(l)):
    print("."+l[i],end="")
```

### 22359: Goldbach Conjecture

http://cs101.openjudge.cn/practice/22359/


1-nin-i




```python
n=int(input())
prime=[1]*(n+1)
prime[0]=prime[1]=0
for i in range(2,int(n**0.5)+1):
    if prime[i]:
        for j in range(i**2,n+1,i):
            prime[j]=0
for i in range(n+1):
    if prime[i] and prime[n-i]:
        print(i,end=' ')
        print(n-i)
        break
```



 <mark>"Accepted"</mark>

```python
l=input().split('+')
m=0
for i in range(len(l)):
    li=len(l[i])
    for j in range(li):
        if l[i][j]=='n':
            if j==0:
                a=1
            else:
                a=int(l[i][:j])
            b=int(l[i][j+2:])
    if a!=0 and b>m:
        m=b
print('n^'+str(m))
```

### 24684: 

http://cs101.openjudge.cn/practice/24684/







```python
from collections import defaultdict
l=list(map(int,input().split()))
dic=defaultdict(int)
for i in range(len(l)):
    dic[l[i]]+=1
l1=[]
for key in dic:
    l1.append([dic[key],key])
l1.sort(key=lambda x:(-x[0],x[1]))
i=0
ans=[]
while i<len(l1) and l1[i][0]==l1[0][0]:
    ans.append(l1[i][1])
    i+=1
print(*ans)

```
